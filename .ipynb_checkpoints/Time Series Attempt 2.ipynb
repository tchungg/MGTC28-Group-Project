```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
df= pd.read_csv('Total Sales N3.csv')
```

```python
print(df.head())
```

```python
#finding rows with zeros in any column
rows_with_zeros = df[(df == 0).any(axis=1)]
# Remove rows with zeros
df_no_zeros = df[(df != 0).all(axis=1)]
# Checking to make sure rows were removed
df_no_zeros
df=df_no_zeros
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# Set 'Date' as the index
df.set_index('Date', inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Amount'])
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df['Amount'])
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df['Amount'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
