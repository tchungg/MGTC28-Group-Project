```python

import pandas as pd

# Load the dataset
file_path = '~/Downloads/kenvue_data/Cleaned_Customer_DC_Inventory.csv'  
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
        
```

```python

# Convert the 'Date' column to datetime type for time series analysis
data['Date'] = pd.to_datetime(data['Date'])
        
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Aggregate data by Date
aggregated_data = data.groupby('Date').agg({'Dc Amount': 'sum', 'Store Amt On Hand': 'sum'}).reset_index()

# Plotting the trends
plt.figure(figsize=(15, 6))
sns.lineplot(x='Date', y='Dc Amount', data=aggregated_data, label='DC Amount')
sns.lineplot(x='Date', y='Store Amt On Hand', data=aggregated_data, label='Store Amount On Hand')
plt.title('Seasonal Trends in DC Amount and Store Amount On Hand')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
        
```
