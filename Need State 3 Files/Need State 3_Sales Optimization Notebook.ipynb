```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Load the data
sales_data = pd.read_csv('~/Downloads/Kenvue_data/Need State 3_Sales Optimization.csv')
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data.set_index('Date', inplace=True)
sales_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
print(sales_data.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(sales_data['Amount_Factory_POS'], label='Factory POS Sales')
plt.plot(sales_data['Amount_Total_Ecomm_POS'], label='E-commerce POS Sales')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.legend()
plt.show()
```

```python
# Yearly sales analysis
yearly_sales = sales_data.resample('Y').sum()
print('Yearly Sales:\n', yearly_sales)

# Weekly sales analysis
weekly_sales = sales_data.resample('W').sum()
print('Weekly Sales:\n', weekly_sales)
```
